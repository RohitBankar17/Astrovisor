```python
#importing Pandas Library
import pandas as pd
```

```python
#reading Astrosat Publications.txt file into dataframe using pandas
df2=pd.read_csv('AS_publications2019-21_m2.txt',sep='\n',names=['data'],engine='python')
```

```python
df2.head()
```

```python
# Segregrating Heading and Descriptions Seperately
df2[['Heading','Description']]=df2['data'].str.split(":",expand=True)
```

```python
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
df2
```

```python
#Dropping unnecessary column
df2.drop('data',axis=1,inplace=True)
```

```python
df2
```

```python
df2.tail(20)
```

```python
# Creating List corresponding to Title, Authors, Bibliographic Codes, Keywords, Abstract and URL
Title=[]
Authors=[]
Bibliographic_Code=[]
Keywords=[]
Abstract=[]
URL=[]
```

```python
# Shape of DataFrame
df2.shape[0]
```

```python
# Differentiating DataFrame into Title, Authors, Bibliographic Codes, Keywords, Abstract and URL

for ind in range(0,df2.shape[0]):
    if((ind%6)==0):
        Title.append(df2.iloc[ind,1])
    
    if((ind%6)==1):
        Authors.append(df2.iloc[ind,1])
    
    if((ind%6)==2):
        Bibliographic_Code.append(df2.iloc[ind,1])
        
    if((ind%6)==3):
        Keywords.append(df2.iloc[ind,1])
        
    if((ind%6)==4):
        Abstract.append(df2.iloc[ind,1])
    
    if((ind%6)==5):
        URL.append(df2.iloc[ind,1])
        
```

```python
# Creation of Final DataFrame
dff=pd.DataFrame()
```

```python
dff['Title']=Title
```

```python
dff['Authors']=Authors
```

```python
dff['Bibliographic_Code']=Bibliographic_Code
```

```python
dff['Keywords']=Keywords
```

```python
dff['Abstract']=Abstract
```

```python
dff['URL']=URL
```
