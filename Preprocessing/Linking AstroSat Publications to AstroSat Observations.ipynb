```python
# Getting preprocessed Astrosat publications(Catalog C) in dfp dataframe
dfp=pd.read_csv('Modified_AS_publications_1.csv')
```

```python
#Seperate list for Bibliographic _Codes
Bibliographic_Codes=[]
```

Searching for name from dfo in dfp to link Catalog B and C.

```python
# Search algorithm to link Catalog B and Catalog C
# It takes names of the object from Catalog B(AstroSat observations) and Search in Catalog C(Astrosat Publications)
# It returns Bibliographic Codes of the research publications associated with the object
for i in range(0,dfo.shape[0]):
    name=dfo['Name'][i]
    res = [i for j in str(name).split() for i in (j, ' ')][:-1]
    res=list(set(res))
    res=[s for s in res if s != ' ']
    st=""
    for nn in res:
        for j in range(0,dfp.shape[0]):
            if(nn in str(dfp['Abstract'][j])):
                st=st+str(dfp['Bibliographic_Code'][j])+","
            elif (nn in str(dfp['Title'][j])):
                st=st+str(dfp['Bibliographic_Code'][j])+","
            elif (nn in str(dfp['Keywords'][j])):
                st=st+str(dfp['Bibliographic_Code'][j])+","
    if(len(st)==0):
        Bibliographic_Codes.append("-1")
    else:
        Bibliographic_Codes.append(st)
        
```

```python
# Checking Dimensions
print(len(Bibliographic_Codes))
```

```python
#Adding Bibliographic Codes to our data frame dfo.
dfo['Bibliographic_Codes']=Bibliographic_Codes
```
