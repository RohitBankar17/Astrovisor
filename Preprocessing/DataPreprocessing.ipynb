# Data preprocessing

```python
#Based on the readme files, we located the catalogue in the Vizier database, here we obtain the pertinent files and work on them
from astroquery.vizier import Vizier
v = Vizier(columns = ['all'])
v.ROW_LIMIT = -1
catalog_list = v.find_catalogs('V/106')
catalogs = v.get_catalogs(catalog_list.keys())
```

# Table 1


```python
df = catalogs[0].to_pandas() #Converting the catalogs file to a dataframe
```

```python
df.head()
```

```python
dfToAppend = catalogs[4].to_pandas() #The catalog containing a list of authors and references
```

```python
dfToAppend.head()
```

```python
listRef = [] #Obtaining a column of the authors/bibCodes based on the references mentioned in the catalog
for i in range(len(df)):
#     print(df.iloc[i, 11],df.iloc[i, 24], df.iloc[i, 29], df.iloc[i, 35])
    temp = df.iloc[i, 35].decode()+','+df.iloc[i, 24].decode()+','+ df.iloc[i, 29].decode()+','+df.iloc[i, 35].decode()
    listS = set(temp.split(','))
    listS.discard('')
    listing=''
    for val in listS:
        for j in range(len(dfToAppend)):
            if val == str(dfToAppend['Ref'][j])+ dfToAppend['m_Ref'][j].decode():
                listing+= dfToAppend['BibCode'][j].decode()+ '-'+ dfToAppend['Aut'][j].decode()+','
                #format: Bibcode1-Author1, Bibcode2-Author2, ...
    listRef.append(listing)
    
    
                
    
    
```

```python
# df.drop(['BibCodes'], axis = 1)
```

```python
df['BibCodes'] = listRef #Adding a column to the existing dataframe for bibcode and authors
```

```python
df.tail()
```

```python
from astropy.table import Table
table = Table.from_pandas(df) #converting the dataframe to pandas
```
